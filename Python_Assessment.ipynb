```python
# Import the houses.csv file

file_path = 'houses.csv'

df_house = pd.read_csv(file_path, sep= ",")

display(df_house)
```

```python
# Replace blank/missing values in column "Garden" with value "NULL"

df_house.isna().sum()
```

```python
df_house["Garden"].fillna("NULL", inplace=True)
df_house.isna().sum()
```

```python
# Remove all rows for houses that were not sold (where Sold_flag = "No")

df_house = df_house[df_house["Sold_flag"] != "No"]
display(df_house)
```

```python
# Sort the dataset by Transaction_date and Price in ascending order 
df_house = df_house.sort_values(by=["Transaction_ID", "Price"], ascending=[True, True])
display(df_house)
```

```python
# Change Transaction_date column format like: 29-Aug-2022
df_house["Transaction_date"] = pd.to_datetime(df_house["Transaction_date"], format='ISO8601')
df_house["Transaction_date"] = df_house["Transaction_date"].dt.strftime('%d-%b-%Y')
display(df_house)
```

```python
# In column Real_Estate_agent replace “ABC Properties” as “XYZ Properties”

df_house["Real_estate_agent"] = df_house["Real_estate_agent"].replace("ABC Properties", "XYZ Properties")
display(df_house)
```

```python
# Calculate the average price of all houses and store the value in column "Price_avg"

df_house["Prive_avg"] = df_house["Price"].mean()
display(df_house)
```

```python
# Create a new column with a boolean to indicate whether "Price" is below the average "Price_avg" (If Price < Price_avg then True and If Price >= Price_avg then False)
import numpy as np
df_house["Avg_Plag"] = np.where(df_house["Price"] < df_house["Prive_avg"], True, False)
display(df_house)
```

```python
# Remove all digits in column "Market" to clean data and to obtain two possible values: "Secondary" and  "Primary" 
df_house['Market'] = df_house['Market'].replace(to_replace=r'\d', value='', regex=True)
df_house['Market'].unique()
```

```python
# Segregate houses based on Size_meters column into following buckets and store the result in new column called “House_Size”:
		# a.	Small: 25 – 50
		# b.	Medium: 51 – 100 
		# c.	Large: 101 - 200
		# d.	XLarge: 200+

```

```python
# Create a pivot table/overview that shows the number of houses sold per city and market type. Store the pivot table as a separate dataframe

pivot_df = df_house.pivot_table(index='City', columns='Market', values='House_ID', aggfunc='count')
display(pivot_df)
```

```python
# Use Price as the dependent variable and fit a simple linear regression with an independent variable of your choice. Interpret the results
import statsmodels.api as sm

X = df_house['Size_meters']
y = df_house['Price']

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()

print(model.summary())
```

```python
# Import the second data set (transactions.csv) 

file_path = 'transactions.csv'

df_trans = pd.read_csv(file_path, sep= ",")

display(df_trans)
```

```python
# Merge two datasets in such a way that you add column "Final_price" from Transaction dataset to the main dataset houses

df_house_2 = pd.merge(df_house, df_trans, on="Transaction_ID" )

display(df_house_2)
```

```python
# Calculate the difference between the price and the final price but only for the secondary market. For the primary market paste value "NA"

df_house_2['Price_Difference'] = df_house_2.apply(lambda row: row['Final_price'] - row['Price_x'] if row['Market'] == 'Secondary' else 'NA', axis=1)

display(df_house_2)
```

```python
# Plot the following graph of your choice: On x_axis we should see City and on y_axis we should see the average price of houses sold in that city and the count of houses sold in that city
import matplotlib.pyplot as plt

df_house_sum = df_house.groupby('City').agg({'Price': ['mean', 'count']}).reset_index()
df_house_sum.columns = ['City', 'Average_Price', 'Count']

plt.figure(figsize=(10, 6))
plt.bar(df_house_sum['City'], df_house_sum['Average_Price'])
plt.ylabel('Average Price')
plt.xlabel('City')
plt.title('Average Price of Houses Sold by City')
plt.xticks(rotation=45)

ax2 = plt.gca().twinx()
df_house_sum['Count'].plot(ax=ax2, color='red', marker='o', linewidth=2)
ax2.set_ylabel('Count')

plt.show()
```
